```python
import pandas as pd
import matplotlib.pyplot as plt

cap_data = pd.read_csv("repriskdata_2400_us_comp_returns.csv")
print(cap_data.head())
```

```python
cap_data["year"] = pd.DatetimeIndex(cap_data['date']).year
cap_data["month"] = pd.DatetimeIndex(cap_data['date']).month
print(cap_data["year"].head())
print(cap_data["month"].head())
print(cap_data.shape)

cap_data["year"].dropna()
cap_data["month"].dropna()
print(cap_data.shape)
```

```python
cap_10to20 = cap_data[cap_data["year"] >= 2010]
print(cap_10to20.head())
```

```python
years = cap_10to20["year"].unique()
months = cap_10to20["month"].unique()

#esg good minus bad factor
egmb_factor = pd.DataFrame({'year' : [],'month':[],'factor':[],'quantile':[],'mean':[]})
esg_quantile = cap_10to20["current_rri"].quantile(.5)

cap_10to20 = cap_10to20.reset_index()
cap_10to20.monthly_return = cap_10to20.monthly_return.fillna(0.0)
returns_not_reported =[]
for i in range(len(cap_10to20.monthly_return)):
    if cap_10to20.monthly_return[i] == 'fail':
        returns_not_reported.append(i)
cap_10to20 = cap_10to20.drop(returns_not_reported)
cap_10to20.monthly_return = cap_10to20.monthly_return.astype(float) 

val_checker = cap_10to20.sort_values("monthly_return",ascending= False)
print(val_checker.head(30))

checker_data = cap_10to20[(cap_10to20["year"] == 2011) & (cap_10to20["month"] == 5)]


#plt.scatter(checker_data["current_rri"],checker_data["monthly_return"])
#plt.show()
```

```python
row_median = checker_data[checker_data['current_rri']==checker_data['current_rri'].quantile(interpolation='nearest')]

checker_data = checker_data.sort_values(["monthly_return"],ascending=[False])
checker_data = checker_data.reset_index()
print(checker_data.head(50))
```

```python
for i in years:
    for j in months:
        temp_data = cap_10to20[(cap_10to20["year"] == i) & (cap_10to20["month"] == j)]
        esg_quantile = temp_data["current_rri"].quantile(.5)
        mean_val = temp_data["current_rri"].mean()
        return_mean = temp_data["monthly_return"].mean()
        val = 0
        temp_data = temp_data.sort_values(["current_rri","country_sector_average"],ascending= [True,True])
        temp_data = temp_data.reset_index()
        for k in range(len(temp_data["current_rri"])):
            if (k <= len(temp_data["current_rri"])/2) & (abs(temp_data["monthly_return"].iloc[k]) < 10) :
                val = val + float(temp_data["monthly_return"].iloc[k])   
            elif (k > len(temp_data["current_rri"])/2) & (abs(temp_data["monthly_return"].iloc[k]) < 10):
                val = val - float(temp_data["monthly_return"].iloc[k])
        df2 = {'year':i,'month':j,'factor':val,'quantile':esg_quantile,'mean':mean_val}
        egmb_factor = pd.concat([egmb_factor, pd.DataFrame.from_records([df2])])

print(egmb_factor.head(40))
```
